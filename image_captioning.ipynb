Here we are, in the Image Captioning task

```python
import os
import matplotlib.pyplot as plt
```

Firstly, get the files count

```python
train_dir = './train2014'
val_dir = './val2014'
test_dir = './test2014'
flickr_dir = './flickr'

file_numbers = {}

file_numbers['Coco_train'] = len([file for file in os.listdir(train_dir) if os.path.isfile(os.path.join(train_dir, file))])
file_numbers['Coco_val'] = len([file for file in os.listdir(val_dir) if os.path.isfile(os.path.join(val_dir, file))])
file_numbers['Coco_test'] = len([file for file in os.listdir(test_dir) if os.path.isfile(os.path.join(test_dir, file))])
file_numbers['Flickr'] = len([file for file in os.listdir(flickr_dir) if os.path.isfile(os.path.join(flickr_dir, file))])

print(file_numbers)
plt.figure(figsize=(12, 8))
plt.title('Pictures number', fontdict={'size': 24})
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.bar(file_numbers.keys(), file_numbers.values())
plt.show()
```
